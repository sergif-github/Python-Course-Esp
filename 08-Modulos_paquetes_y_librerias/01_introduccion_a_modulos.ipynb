# Módulo 8: Módulos, paquetes y librerías

## Parte 1: Introducción a los módulos

En Python, un módulo es un archivo que contiene código de Python que define funciones, clases y variables. Los módulos proporcionan una manera
para organizar y reutilizar el código, lo que facilita la gestión de programas grandes y la colaboración con otros. Al importar módulos,
puede acceder a la funcionalidad definida dentro de ellos y aprovechar sus capacidades en su propio código.
Por otro lado, un paquete es una forma de organizar módulos relacionados en una jerarquía de directorios y una biblioteca es una colección de módulos o paquetes que proporcionan código preescrito y funcionalidad para realizar tareas específicas o resolver problemas comunes.

### 1.1. Usando módulos

Para usar un módulo en su programa Python, primero debe importarlo. Python proporciona varias formas de importar módulos:

- Importación de todo el módulo

   Puede importar todo el módulo utilizando la declaración de importación seguida del nombre del módulo. Por ejemplo:

     ```python
     import math
     ```

   Una vez importado, puede acceder a las funciones, clases y variables definidas en el módulo utilizando el nombre del módulo como prefijo. Por ejemplo, para acceder a la función sqrt() desde el módulo matemático, escribiría math.sqrt().

- Importación de artículos específicos

   Si solo necesita funciones, clases o variables específicas de un módulo, puede importarlas directamente usando la palabra clave from. Por ejemplo:

     ```python
     from math import sqrt, pi
     ```

   Con esta declaración de importación, puede usar directamente sqrt() y pi en su código sin necesidad de prefijarlos con el nombre del módulo.

- Importar con un alias

   También puede importar un módulo o elementos específicos con un alias utilizando la palabra clave as. Esto puede ser
útil para proporcionar nombres más cortos o más significativos.

   Por ejemplo:

     ```python
     import math as m
     ```

   Con esta declaración de importación, puede usar m.sqrt() en lugar de math.sqrt().

### 1.2. Módulos de biblioteca estándar

Las bibliotecas pueden ser parte de la biblioteca estándar de Python, que se incluye con Python mismo, o bibliotecas de terceros creadas por otros desarrolladores.
Python viene con un amplio conjunto de bibliotecas estándar que brindan una amplia gama de funcionalidades para diversas tareas.
Estos módulos se incluyen con Python y se pueden usar fácilmente sin necesidad de una instalación adicional. Proporcionan una amplia gama de funcionalidades, lo que facilita a los desarrolladores realizar tareas comunes sin tener que construir todo desde cero.

Algunos de los módulos de biblioteca estándar más utilizados incluyen:

- random:

   Este módulo proporciona funciones para generar números aleatorios, seleccionar elementos aleatorios de una secuencia y mezclar secuencias.

- datetime

   El módulo de fecha y hora permite la manipulación y el formato de fechas y horas. Proporciona clases para trabajar con fechas, horas, zonas horarias e intervalos.

- os

   El módulo os proporciona una forma de interactuar con el sistema operativo. Le permite realizar diversas tareas relacionadas con el sistema operativo, como trabajar con archivos y directorios, administrar procesos y acceder a variables de entorno.

- csv

   El módulo csv proporciona funcionalidad para leer y escribir archivos CSV (valores delimitados). Le permite manejar fácilmente datos tabulares en un formato estructurado.

- math
  
   Proporciona funciones matemáticas y constantes.

- json
  
   Ofrece funciones para trabajar con datos JSON.

Estos son solo algunos ejemplos, y hay muchos más módulos de biblioteca estándar disponibles para varios propósitos, incluida la programación de redes (socket), expresiones regulares (re), serialización de datos (pickle) y más.

Para explorar la lista completa de módulos de biblioteca estándar disponibles y sus funcionalidades, puede consultar la documentación oficial de Python en https://docs.python.org/3/library/index.html.

### 1.3. Módulos de terceros

Además de los módulos de la biblioteca estándar, Python tiene un vasto ecosistema de módulos de terceros que se pueden instalar y
utilizado en sus proyectos. Los módulos de terceros son creados por la comunidad de Python y ofrecen funcionalidad adicional para aplicaciones específicas.
propósitos como desarrollo web, análisis de datos, aprendizaje automático y más. Algunos módulos populares de terceros incluyen numpy,
pandas o matplotlib. Estos módulos se pueden instalar usando administradores de paquetes como pip o conda.

### 1.4. Resumen

Los módulos son una parte esencial de la programación de Python, lo que le permite organizar y reutilizar el código. Al importar módulos, puede
acceder a funciones, clases y variables definidas en ellas, lo que facilita el aprovechamiento de la funcionalidad existente y la creación de más
aplicaciones poderosas. Python proporciona varias formas de importar módulos, lo que le permite importar el módulo completo, elementos específicos,
o asignar alias. Además, Python ofrece un amplio conjunto de módulos de biblioteca estándar y una amplia colección de módulos de terceros.
para ampliar las capacidades de sus programas.
