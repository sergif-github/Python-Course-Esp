### 1.1. ¿Qué es Python?

Python es un lenguaje de programación interpretado de alto nivel conocido por su simplicidad y legibilidad.
Fue creado por Guido van Rossum y lanzado por primera vez en 1991. La filosofía de diseño de Python enfatiza la legibilidad del código y se enfoca en minimizar el esfuerzo requerido para escribir y mantener el código.

Python ofrece una sintaxis limpia y elegante que permite a los programadores expresar conceptos en menos líneas de código en comparación
a otros lenguajes de programación. Esta simplicidad y legibilidad hacen de Python una excelente opción tanto para principiantes como para desarrolladores experimentados.

El lenguaje proporciona una amplia gama de estructuras de datos integradas y admite varios paradigmas de programación, incluidos procedimientos,
programación orientada a objetos y funcional. También ofrece una amplia colección de bibliotecas y marcos de terceros,
haciéndolo versátil y adecuado para varios dominios de aplicación.

La popularidad de Python ha crecido significativamente a lo largo de los años, impulsada por su naturaleza fácil de usar, el amplio apoyo de la comunidad,
y su aplicabilidad a una amplia gama de campos. Es ampliamente utilizado en desarrollo web, análisis de datos, computación científica,
inteligencia artificial, aprendizaje automático, automatización y más.

El éxito de Python se puede atribuir a su extensa biblioteca estándar, que proporciona un amplio conjunto de módulos y herramientas para
varias tareas Además, la comunidad de Python contribuye activamente a un vasto ecosistema de bibliotecas y marcos de código abierto,
ampliando aún más sus capacidades.

Ya sea que sea un programador principiante o experimentado, aprender Python puede capacitarlo para abordar diversos proyectos,
automatice tareas repetitivas, analice datos, cree aplicaciones web y desarrolle tecnologías de vanguardia.


### 1.2. Historia y antecedentes de Python

Python tiene una historia interesante que abarca más de tres décadas. Fue creado por Guido van Rossum, un programador holandés,
a fines de la década de 1980. Guido desarrolló Python como un proyecto de pasatiempo mientras trabajaba en el Instituto Nacional de Investigación para
Matemáticas y Ciencias de la Computación en los Países Bajos. Su objetivo era diseñar un lenguaje que fuera intuitivo, fácil de leer,
y enfatizó la legibilidad del código.

El nombre "Python" se inspiró en el amor de Guido por la serie de comedia británica "Monty Python's Flying Circus". Guido quería un corto,
nombre único y ligeramente misterioso que destacaría en el panorama de los lenguajes de programación. El nombre también refleja la comunidad
el humor y la naturaleza amante de la diversión que se ha convertido en sinónimo de Python.

El primer lanzamiento público de Python, Python 0.9.0, se produjo en febrero de 1991. Desde el principio, Python llamó la atención por su simplicidad y elegancia, atrayendo rápidamente a una base de usuarios devotos. Guido continuó liderando el desarrollo de Python supervisando el crecimiento del lenguaje y asegurando que sus principios básicos permanecieran intactos.

Python 2, lanzado en 2000, introdujo varias mejoras en el lenguaje. Sin embargo, con el tiempo se convirtió en
evidente que el lenguaje necesitaba algunos cambios fundamentales para abordar ciertos defectos y limitaciones de diseño. Esto condujo a la
desarrollo de Python 3, con Guido van Rossum y la comunidad de Python trabajando juntos para crear una versión compatible con versiones anteriores,
versión aún más moderna y eficiente de Python.

Python 3 se lanzó en 2008 y trajo mejoras significativas al lenguaje. Introdujo características tales como
compatibilidad con Unicode, mejor manejo de cadenas de bytes, mejoras de sintaxis y un enfoque en la escritura de código más eficiente.
Sin embargo, debido a problemas de compatibilidad con versiones anteriores, la adopción de Python 3 fue inicialmente lenta.

Para facilitar la transición de Python 2 a Python 3, se estableció un cronograma de obsolescencia, con Python 2 alcanzando
su fin de vida el 1 de enero de 2020. Esto alentó a los desarrolladores a migrar sus bases de código a Python 3 y adoptar
las modernas características y mejoras que ofrecía.

El éxito de Python se puede atribuir no solo a sus cualidades técnicas, sino también a su comunidad vibrante e inclusiva.
La comunidad de Python es conocida por su naturaleza colaborativa y acogedora, proporcionando una amplia documentación, bibliotecas,
frameworks y soporte a través de foros en línea, conferencias y grupos de usuarios.

Hoy, Python es uno de los lenguajes de programación más populares en todo el mundo. Sigue evolucionando y adaptándose a los nuevos retos para
permanecer a la vanguardia de los avances tecnológicos en campos como el desarrollo web, la ciencia de datos, el aprendizaje automático y la inteligencia artificial.

### 1.4. Python 2 frente a Python 3

Python ha experimentado una transición significativa de Python 2 a Python 3. Python 2 era la versión dominante del lenguaje durante muchos años, pero con el lanzamiento de Python 3, la comunidad de Python hizo un esfuerzo consciente para mejorar el lenguaje y abordar ciertos defectos y limitaciones de diseño.

Diferencias clave
1. Declaración de impresión frente a función de impresión

     En Python 2, la impresión en la consola se realizaba mediante la declaración de impresión "print", mientras que
Python 3 introdujo la función print(). Este cambio permite más flexibilidad y consistencia, ya que la función print()
se comporta como cualquier otra función en Python.

2. Compatibilidad con Unicode

    Python 2 trató las cadenas como una secuencia de bytes de forma predeterminada, mientras que Python 3 convirtió a Unicode en el tipo de cadena predeterminado.
    Este cambio asegura un mejor soporte para la internacionalización y las aplicaciones multilingües, lo que facilita su manejo.
    codificaciones de caracteres diferentes.

3. Operador de División

     En Python 2, el operador de división / entre dos enteros realizaría la división de enteros, devolviendo solo el cociente.
Sin embargo, en Python 3, el operador de división realiza una verdadera división y devuelve un flotante. Para lograr la división de enteros en Python 3,
puede usar el operador //.

4. Iteración y bucle

     La función range() en Python 3 devuelve un iterador en lugar de una lista como lo hace Python 2s, lo que mejora el rendimiento
y uso de memoria para rangos grandes.

5. Manejo de excepciones

     En Python 2, la declaración de excepción permitía capturar múltiples excepciones en un solo bloque,
mientras que Python 3 introdujo una sintaxis más explícita que requiere enumerar cada excepción por separado. Este cambio mejora la claridad y la capacidad de mantenimiento del código.


### 1.5. Transición de Python 2 a Python 3

Python 3 se introdujo para abordar limitaciones de Python 2 y mejorar el diseño general del lenguaje. Sin embargo,
debido a problemas de compatibilidad con versiones anteriores, la adopción de Python 3 fue inicialmente lenta. Esto resultó en un período en el que ambos
Python 2 y Python 3 fueron ampliamente utilizados.

Para fomentar la transición de Python 2 a Python 3, se estableció un cronograma de obsolescencia.
Python 2 llegó al final de su vida útil el 1 de enero de 2020 y ya no se mantiene ni actualiza activamente.
Python 3, por otro lado, continúa recibiendo actualizaciones, mejoras y nuevas funciones.

Para nuevos proyectos, se recomienda usar Python 3 para aprovechar sus características mejoradas, correcciones de errores y el mayor
ecosistema de bibliotecas y herramientas compatibles con Python 3. Sin embargo, para proyectos heredados o sistemas que aún dependen de Python 2,
hay herramientas disponibles, como 2to3, que pueden ayudar a migrar el código a Python 3.

Python 3 ofrece numerosos beneficios y mejoras sobre Python 2, lo que la convierte en la versión preferida para nuevos desarrollos.
A medida que la comunidad de Python continúa evolucionando y adoptando Python 3, es esencial comprender las diferencias y los beneficios
para tomar decisiones informadas sobre el uso de la versión adecuada de Python para sus proyectos.
